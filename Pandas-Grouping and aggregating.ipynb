```python
#Method-2 - to find the percentage of people in each country responded knows Python

#Step-1 : find total resp from country
total_resp = df['Country'].value_counts()
#print(total_resp)
#print()
#Step-2 : find the resp that knows python
Py_resp = country_grp['LanguageWorkedWith'].apply(lambda x: x.str.contains('Python').sum())
#print(Py_resp)

#step-3: concat the above two based on the matching columns

py_df = pd.concat([total_resp, Py_resp], axis ='columns', sort=False)
py_df.rename(columns={'Country': 'Total', 'LanguageWorkedWith': 'Py_users'}, inplace=True)
py_df['Py_percent'] = (py_df['Py_users']/py_df['Total'])*100
py_df
```

```python
from matplotlib import pyplot as plt

from collections import Counter
```

```python
lang_resp = df['LanguageWorkedWith']
lang_resp.fillna(' ', inplace=True)
```

```python
lang_counter = Counter()

for resp in lang_resp:
    lang_counter.update(resp.split(';'))
    
print(lang_counter)

lang=[]
resp=[]

for item in lang_counter.most_common(15):
    lang.append(item[0])
    resp.append(item[1])
    
print(lang.reverse())
resp.reverse()
```

```python
plt.barh(lang, resp)
```
